### 1-D Scatter Plot

```python
sns.set_style("whitegrid")

one=df.loc[df['survival_status']==1]
two=df.loc[df['survival_status']==2]

plt.plot(one['age'],np.zeros_like(one['age']),'o')
plt.plot(two['age'],np.zeros_like(two['age']),'o')

plt.title('1-D Scatter Plot for age vs survival status')
plt.xlabel('age')
plt.ylabel('survival status')
plt.show()
```

OBSERVATION(S):

- Death rate is high for people of age between 41-70

### 2-D Scatter Plot

```python
df.plot(kind='scatter', x='age', y='lymph_nodes') ;
plt.show()

df.plot(kind='scatter', x='age', y='year_of_treatment') ;
plt.show()
```

```python
# 2-D Scatter plot with color-coding for each class.

sns.set_style("whitegrid")
sns.FacetGrid(df,hue='survival_status',size=5) \
   .map(plt.scatter,'age','lymph_nodes') \
   .add_legend()
plt.title('2-D Scatter Plot for age vs lymph nodes')
plt.show()


sns.set_style("whitegrid")
sns.FacetGrid(df,hue='survival_status',size=5) \
   .map(plt.scatter,'age','year_of_treatment') \
   .add_legend()
plt.title('2-D Scatter Plot for age vs year of treatment')
plt.show()
```

# PDF

# CDF

```python
counts,bin_edges = np.histogram(one['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges = np.histogram(two['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('pdf & cdf of age')
plt.xlabel('age')
plt.show();
```

```python
counts,bin_edges = np.histogram(one['year_of_treatment'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges = np.histogram(two['year_of_treatment'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('pdf & cdf of year_of_treatment')
plt.xlabel('year of treatment')
plt.show();
```

```python
counts,bin_edges = np.histogram(one['lymph_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges = np.histogram(two['lymph_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('pdf & cdf of lymph nodes')
plt.xlabel('lymph nodes')
plt.show();
```

# Mean and Std-dev

```python
print('Means:')
print(np.mean(df['age']))
#mean with an outlier
print(np.mean(np.append(df['age'],50)))
print(np.mean(df['year_of_treatment']))
print(np.mean(df['lymph_nodes']))
print(np.mean(df['survival_status']))

print()
print('Std-dev:')
print(np.std(df['age']))
print(np.std(df['year_of_treatment']))
print(np.std(df['lymph_nodes']))
print(np.std(df['survival_status']))
```

# Median, Percentile, Quantile and MAD

```python
print('Medians:')
print(np.median(df['age']))
#median with an outlier
print(np.median(np.append(df['age'],50)))
print(np.median(df['year_of_treatment']))
print(np.median(df['lymph_nodes']))
print(np.median(df['survival_status']))

print()
print('Quantiles:')
print(np.percentile(df['age'],np.arange(0,100,25)))
print(np.percentile(df['year_of_treatment'],np.arange(0,100,25)))
print(np.percentile(df['lymph_nodes'],np.arange(0,100,25)))
print(np.percentile(df['survival_status'],np.arange(0,100,25)))

print()
print('Percentile:')
print(np.percentile(df['age'],90))
print(np.percentile(df['year_of_treatment'],90))
print(np.percentile(df['lymph_nodes'],90))
print(np.percentile(df['survival_status'],90))

print()
from statsmodels import robust
print('Median Absolute Deviation:')
print(robust.mad(df['age']))
print(robust.mad(df['year_of_treatment']))
print(robust.mad(df['lymph_nodes']))
print(robust.mad(df['survival_status']))
```
